```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Cybersecurity_Threats.csv")

# Check structure
print(df.info())

# Preview data
df.head()
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
df.groupby('Year').size().plot(kind='line', marker='o', title="Cybersecurity Incidents Per Year")
plt.show()
```

```python
plt.figure(figsize=(12,6))
df['Attack Type'].value_counts().plot(kind='barh', title="Most Frequent Cyber Threats")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df['Target Industry'].value_counts().plot(kind='bar', title="Most Targeted Industries")
plt.show()
```

```python
top_countries = df['Country'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title("Top 10 Most Affected Countries")
plt.xlabel("Number of Cyber Incidents")
plt.show()
```

```python
df.groupby('Year')['Financial Loss (in Million $)'].sum().plot(kind='bar', title="Financial Impact of Cybercrime ($M)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Incident Resolution Time (in Hours)'], bins=30, kde=True)
plt.title("Distribution of Response Times (Hours)")
plt.show()
```
